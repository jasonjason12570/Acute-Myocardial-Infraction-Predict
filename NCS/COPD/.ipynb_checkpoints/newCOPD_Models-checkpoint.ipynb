#### Resampleling imbalanced data
+ There are several ways of resampling imbalanced data to make the model well trained.
  * Random under-sampling
  * Random over-sampling
  * NearMiss

```python
from sklearn.utils import resample
```

```python
### first we concatnate x_tran and y_train
X = pd.concat([x_train, y_train], axis=1)
```

```python
# separate minority and majority classes
alive = X[X.HOSPITAL_MORTALITY==0]
dead = X[X.HOSPITAL_MORTALITY==1]
```

```python
# upsample minority
dead_upsampled = resample(dead,
                          replace=True, # sample with replacement
                          n_samples=len(alive), # match number in majority class
                          random_state=27) # reproducible results
```

```python
# combine majority and upsampled minority
upsampled = pd.concat([alive, dead_upsampled])

# check new class counts
upsampled.HOSPITAL_MORTALITY.value_counts()
```

#### Train the XGBoost Classifier

tunning: https://www.analyticsvidhya.com/blog/2016/03/complete-guide-parameter-tuning-xgboost-with-codes-python/

```python
def CreateBalancedSampleWeights(y_train, largest_class_weight_coef):
    classes = np.unique(y_train, axis = 0)
    classes.sort()
    class_samples = np.bincount(y_train)
    total_samples = class_samples.sum()
    n_classes = len(class_samples)
    weights = total_samples / (n_classes * class_samples * 1.0)
    class_weight_dict = {key : value for (key, value) in zip(classes, weights)}
    class_weight_dict[classes[1]] = class_weight_dict[classes[1]] * largest_class_weight_coef
    sample_weights = [class_weight_dict[y] for y in y_train]
    return sample_weights
```

```python
largest_class_weight_coef = max(df['HOSPITAL_MORTALITY'].value_counts().values)/df.shape[0]
    
#pass y_train as numpy array
weight = CreateBalancedSampleWeights(y_train, largest_class_weight_coef)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
r_auc = roc_auc_score(y_test, r_probs)
# clf_auc = roc_auc_score(y_test, clf_probs)
knn_auc = roc_auc_score(y_test, knn_probs)
logit_auc = roc_auc_score(y_test, logit_probs)
# sv_auc = roc_auc_score(y_test, sv_clf_probs)
rnf_auc = roc_auc_score(y_test, rnf_probs)
xgb_auc = roc_auc_score(y_test, xgb_clf_probs)
# cat_auc = roc_auc_score(y_test, catboost_probs)
```

#### Display the AUROC Scores

```python
print('Random Prediction: AUROC = %.3f' %(r_auc))
# print('Decision Tree: AUROC = %.3f' %(clf_auc))
print('KNearest Neighbor: AUROC = %.3f' %(knn_auc))
print('Logistic Regression: AUROC = %.3f' %(logit_auc))
# print('Support Vector Machine: AUROC = %.3f' %(sv_auc))
print('Random  Forest: AUROC = %.3f' %(rnf_auc))
print('XGboost: AUROC = %.3f' %(xgb_auc))
# print('CatBoost: AUROC = %.3f' %(cat_auc))
```

#### Calculate the ROC curve

```python
r_fpr, r_tpr, _ = roc_curve(y_test, r_probs)
# clf_fpr, clf_tpr, _ = roc_curve(y_test, clf_probs)
knn_fpr, knn_tpr, _ = roc_curve(y_test, knn_probs)
logit_fpr, logit_tpr, _ = roc_curve(y_test, logit_probs)
# sv_fpr, sv_tpr, _ = roc_curve(y_test, sv_clf_probs)
rnf_fpr, rnf_tpr, _ = roc_curve(y_test, rnf_probs)
xgb_fpr, xgb_tpr, _ = roc_curve(y_test, xgb_clf_probs)
# cat_fpr, cat_tpr, _ = roc_curve(y_test, catboost_probs)
```

#### Plot the curve

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.plot(r_fpr, r_tpr, linestyle='--', label='Base line: (AUROC = %.3f)' %r_auc)
# plt.plot(clf_fpr, clf_tpr, linestyle='--', label='Decision Tree: (AUROC = %.3f)' %clf_auc)
plt.plot(knn_fpr, knn_tpr, linestyle='--', label='KNearest Neighbor: (AUROC = %.3f)' %knn_auc)
plt.plot(logit_fpr, logit_tpr, linestyle='--', label='Logistic Regresion: (AUROC = %.3f)' %logit_auc)
# plt.plot(sv_fpr, sv_tpr, linestyle='--', label='Support Vector Machine: (AUROC = %.3f)' %sv_auc)
plt.plot(rnf_fpr, rnf_tpr, linestyle='--', label='Random Forest: (AUROC = %.3f)' %rnf_auc)
plt.plot(xgb_fpr, xgb_tpr, linestyle='--', label='XGBoost: (AUROC = %.3f)' %xgb_auc)
# plt.plot(cat_fpr, cat_tpr, linestyle='--', label='CatBoost: (AUROC = %.3f)' %cat_auc)

#### Title
plt.title('AUROC comparison for models')

#### Axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

#### show legend
plt.legend()
plt.show()
```
